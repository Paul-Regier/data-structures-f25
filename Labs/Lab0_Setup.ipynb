```python
import sys, importlib.util, platform

print("Python executable:", sys.executable)
print("Python version:", platform.python_version())
print("Platform:", platform.platform())
print("Venv in path?", "venvs" in sys.executable)


def has_module(name):
    spec = importlib.util.find_spec(name)
    return spec is not None


print("ipykernel installed:", has_module("ipykernel"))
print("jupyter installed:", has_module("jupyter"))
print("black installed:", has_module("black"))
print("ruff installed:", has_module("ruff"))
```
