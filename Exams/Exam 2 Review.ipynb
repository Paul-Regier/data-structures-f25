3.  Implement a very small hash map with chaining.

Use:
- an internal list `self.table` of length `m` (start with `m = 5`), where each slot holds a Python list of `(key, value)` pairs.
- a hash function `idx = hash(key) % m`.

Write methods:
- `put(key, value)` – insert or update
- `get(key)` – return value or `None` if missing

```python
class MiniMap:
    def __init__(self, m: int = 5):
        self.table = [[] for _ in range(m)]
        self.n = 0
        self.m = m
        self.alpha = self.n/self.m

    def put(self, key, value):
        # TODO: insert OR update if key already exists
        idx = hash(key) % len(self.table)
        bucket = self.table[idx]
        # Reassignment
        for i, (k,v) in enumerate(bucket):
            if key == k:
                bucket[i] = (key,value)
                return
        # Append if not found
        bucket.append( (key,value) )
        self.n +=1
        self.alpha = self.n/self.m

    def get(self, key):
        # TODO: insert OR update if key already exists
        idx = hash(key) % len(self.table)
        bucket = self.table[idx]
        # Search
        for i, (k,v) in enumerate(bucket):
            if key == k:
                return v
        # If not found, return None:
        return None

    def __str__(self):
        return f"alpha: {self.alpha}: {self.table}"
```

```python
m = MiniMap()
m.put(10, "Hello")
m.put(3, "Good bye!")
m.put(8, "Oops")
m.put(10, "Hi")
m.put(13, "Bo!")

print(m)
print(m.get(13))
print(m.get(100))
```
