**13. (5 pts)** Plot the functions `n`, `n log n`, and `n^2` on the same axes for `n = 1..100`. Use `matplotlib` and include a legend. Label axes clearly.


````python
# work here
import numpy as np
import math
import matplotlib.pyplot as plt


# your plotting code
def plot_funcs(funcs, n_max=2000, title="Growth comparison", logy=False):
    n = np.arange(1, n_max + 1)
    plt.figure(figsize=(7, 4.5))
    for label, f in funcs:
        y = f(n)
        plt.plot(n, y, label=label)
    if logy:
        plt.yscale("log")
    plt.xlabel("n")
    plt.ylabel("f(n)")
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
````
